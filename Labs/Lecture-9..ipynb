```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import csv

# Настройка сервиса
service = Service('/opt/homebrew/bin/chromedriver')  # твой путь
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# URL
base_url = "http://quotes.toscrape.com/js/"

# Список для сохранения результатов
quotes_data = []

# Скрейпим первые 5 страниц
for page in range(1, 6):
    driver.get(f"{base_url}page/{page}/")
    time.sleep(2)  # немного подождать, чтобы JS прогрузился

    quotes = driver.find_elements(By.CLASS_NAME, "quote")

    for quote in quotes:
        text = quote.find_element(By.CLASS_NAME, "text").text.strip()
        author = quote.find_element(By.CLASS_NAME, "author").text.strip()
        tags = [tag.text for tag in quote.find_elements(By.CLASS_NAME, "tag")]

        quotes_data.append([text, author, ", ".join(tags)])

# Закрываем драйвер
driver.quit()

# Сохраняем в CSV
with open("quotes.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Text", "Author", "Tags"])
    writer.writerows(quotes_data)

print("Готово! quotes.csv сохранён.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, WebDriverException
from urllib.parse import urljoin, urlparse
import time

# Настройки
service = Service('/opt/homebrew/bin/chromedriver')
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(service=service, options=options)
wait = WebDriverWait(driver, 15)

# Целевая информация
base_url = "https://innopolis.university/en/"
target_text = "Alaa Aldin Hajjar"
visited = set()
to_visit = set([base_url])

def is_valid_url(href):
    return href and href.startswith(base_url) and not href.startswith("mailto:")

found = False

while to_visit:
    url = to_visit.pop()
    if url in visited:
        continue
    visited.add(url)

    try:
        driver.get(url)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        time.sleep(1.5)

        page_source = driver.page_source.lower()

        if target_text.lower() in page_source:
            print(f"\n✅ Имя '{target_text}' найдено на странице: {url}")
            found = True
            break
        else:
            print(f"❌ Проверено: {url}")

        # Собираем новые ссылки с этой страницы
        links = driver.find_elements(By.TAG_NAME, "a")
        for link in links:
            href = link.get_attribute("href")
            if is_valid_url(href):
                clean_href = href.split("#")[0]
                if clean_href not in visited:
                    to_visit.add(clean_href)

    except TimeoutException:
        print(f"⏰ Таймаут на {url}")
    except WebDriverException as e:
        print(f"⚠️ Ошибка на {url}: {e}")

if not found:
    print("\n❌ Имя не найдено ни на одной из просмотренных страниц.")

driver.quit()
```
